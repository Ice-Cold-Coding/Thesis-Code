```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
view_count = df["View Position"].value_counts()
print(view_count)
```

```python
plt.bar(view_count.index, view_count.values)
```

```python
gender_counts = df['Patient Gender'].value_counts()
```

```python
plt.bar(gender_counts.index, gender_counts.values)

plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Patient Gender Distribution')

plt.show()
```

```python
place_holder_dataframe = {'Total Number': [], 'Label': []}
counting_labels = pd.DataFrame(place_holder_dataframe)

for i in df["Finding Labels"].unique():
    new_row = {'Total Number': df[df["Finding Labels"] == i]["Finding Labels"].count(), 'Label': i}
    new_row_df = pd.DataFrame(new_row, index=[0])
    counting_labels = pd.concat([counting_labels, new_row_df], ignore_index=True)
```

```python
counting_labels
```

```python
plt.pie(x = counting_labels["Total Number"],labels = counting_labels["Label"])

plt.title("Chest X-ray14 Class Distribution", fontsize=10)

plt.figure(figsize=(10,10))

plt.show()
```
